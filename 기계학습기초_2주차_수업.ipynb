# ㅁ Combine Data Sets

## 1. merge()
- 문제1. adf, bdf 데이터 프레임을 x1컬럼을 pd.merge()를 사용하여 합치려고 합니다. how 파라메터를 다양하게 주면서 그 의미를 확인해봅니다.
***
<img src='https://drive.google.com/uc?id=1XYG6XRybqhCVBQKbfGC9VmYzzU5FOr-F'>

```python
adf = pd.DataFrame({"x1": ["A", "B", "C"], "x2": [1, 2, 3]})
adf
```

```python
bdf = pd.DataFrame({"x1": ["A", "B", "D"], "x3": ["T", "F", "T"]})
bdf
```

```python
abdf = pd.merge(adf,bdf,how='left', on='x1')
abdf
```

```python
abdf1= pd.merge(adf,bdf,how='right', on='x1')
abdf1
```

```python
abdf2= pd.merge(adf,bdf, how='inner', on='x1')
abdf2
```

```python
abdf3= pd.merge(adf,bdf,how='outer', on='x1')
abdf3
```

### Q&A) 다음 코드의 의미를 확인해봅니다.
- Method Chaining

```python
ydf = pd.DataFrame({"x1": ["A", "B", "C"], "x2": [1, 2, 3]})
zdf = pd.DataFrame({"x1": ["B", "C", "D"], "x2": [2, 3, 4]})
pd.merge(ydf, zdf, how='outer', indicator=True)\
   .query('_merge == "left_only"') \
   .drop(columns=['_merge'])
```

```python
ydf = pd.DataFrame({"x1": ["A", "B", "C"], "x2": [1, 2, 3]})
ydf
```

```python
zdf = pd.DataFrame({"x1": ["B", "C", "D"], "x2": [2, 3, 4]})
zdf
```
